```python
import pandas as pd
import numpy as np
from google.colab import files

uploaded = files.upload()
data = pd.read_csv("data.csv", encoding='latin1')
data.head()
```

```python
print("Розмір датасету (рядків, стовпців):", data.shape)
```

```python
data.replace(to_replace = " ", value = np.NaN)
```

```python
data.size
```

```python
missing_values = data.isnull().sum()
print("Пропущені значення у кожному стовпці:\n", missing_values)
```

```python
numeric_columns = data.select_dtypes(include=[np.number]).columns
data[numeric_columns] = data[numeric_columns].fillna(data[numeric_columns].mean())
data[numeric_columns] = data[numeric_columns].astype(float)
print(data)
print(data.dtypes)
```

```python
duplicate_rows = data.duplicated().sum()
print("Кількість дублікатів у датасеті:", duplicate_rows)
```

```python
unique_countries = data['Country'].nunique()
print('Кількість унікальних країн:', unique_countries)
```

```python
# Діаграма розподілу переглядів

data['video views'].plot(kind='hist', title='Розподіл переглядів')
```

```python
# Кількість переглядів на Youtube

max_views = data['video views'].max()
min_views = data['video views'].min()
mean_views = data['video views'].mean()

print('найменшою кількістю переглядів -', min_views)
print('найбільшою кількістю переглядів -', max_views)
print('середньою кількістю переглядів -', mean_views)
```

```python
# Найбільше завантажень на Youtube

country_with_most_uploads = data.loc[data['uploads'].idxmax()]['Country']
print("Країна з найбільшою кількістю завантажень на YouTube:", country_with_most_uploads)
```
