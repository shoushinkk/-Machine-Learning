```python
#1.Cтворіть одновимірний масив (вектор) з добутками парних чисел на їхні попередні непарні числа, починаючи з 2, та виведіть його значення.

result = []
#Перше парне число
current_even = 2

for _ in range(10):

    current_odd = current_even - 1
#Множення парного на попереднэ непарне
    product = current_even * current_odd
    result.append(product)
#Наступне парне число
    current_even += 2
#Вивід результату
print(result)
```

```python
#2.Створіть двовимірний масив (матрицю) розміром 3x3, заповніть його значеннями квадратів відстаней між кожною парою натуральних чисел, починаючи з 1, та виведіть його значення.

import numpy as np

# Створюємо матрицю розміром 3x3
matrix = np.zeros((3, 3))

# Заповнюємо матрицю квадратами відстаней
for i in range(3):
    for j in range(3):
        distance = (i+1 - (j+1))**2
        matrix[i][j] = distance

# Виводимо матрицю
print(matrix)
```

```python
#3.Створіть масив розміром 5x5, заповніть його випадковими дійсними числами від -2 до 2 та знайдіть середнє за модулем значення його елементів.
import numpy as np

# Створюємо масив розміром 5x5 з випадковими дійсними числами від -2 до 2
random_array = np.random.uniform(-2, 2, size=(5, 5))

# Знаходимо модуль кожного елемента масиву за допомогою бібліотеки numpy
abs_array = np.abs(random_array)

# Знаходимо середнє значення модулів елементів масиву за допомогою бібліотеки numpy
average_abs_value = np.mean(abs_array)

print("Масив:")
print(random_array)
print("Середнє за модулем значення елементів масиву:", average_abs_value)
```

```python
#4.Створіть масив розміром 5x5, заповніть його випадковими дійсними числами в діапазоні від 0 до 1 та обчисліть суму елементів, які менше середнього значення.
import numpy as np

array = np.random.rand(5, 5)

mean_value = np.mean(array)

less_than_mean = array[array < mean_value]

sum_of_less_than_mean = np.sum(less_than_mean)

print("Масив:")
print(array)
print("\nСереднє значення:", mean_value)
print("\nСума елементів, які менше середнього значення:", sum_of_less_than_mean)
```

```python
#11.Створіть матрицю розміром 2x3 та вектор розміром 3, заповніть матрицю випадковими дійсними числами від 0 до 1, а вектор - випадковими цілими числами в діапазоні від 1 до 10, і знайдіть суму елементів вектору, які більше за відповідний елемент матриці.
import numpy as np

# Створення матриці розміром 2x3 з випадковими дійсними числами від 0 до 1
matrix = np.random.rand(2, 3)

# Створення вектора розміром 3 з випадковими цілими числами в діапазоні від 1 до 10
vector = np.random.randint(1, 11, 3)

# Виведення створеної матриці та вектора
print("Матриця:")
print(matrix)
print("Вектор:")
print(vector)

# Ініціалізуємо змінну для зберігання суми
result = 0

# Порівнюємо кожний елемент вектора з відповідним елементом матриці
for i in range(2):
    for j in range(3):
        if vector[j] > matrix[i][j]:
            result += vector[j]

print("Сума елементів вектору, які більше за відповідний елемент матриці:", result)

```

```python
#12.Створіть дві матриці розміром 4x4, заповніть їх випадковими цілими числами в діапазоні від 1 до 10 та знайдіть добуток елементів діагоналі першої матриці.
import random

# Створюємо першу матрицю та заповнюємо її випадковими числами
matrix1 = [[random.randint(1, 10) for _ in range(4)] for _ in range(4)]
matrix2 = [[random.randint(1, 10) for _ in range(4)] for _ in range(4)]

# Виводимо  матриці
print("Перша матриця:")
for row in matrix1:
    print(row)

print("\nДруга матриця:")
for row in matrix2:
    print(row)


# Обчислюємо добуток елементів діагоналі першої матриці
diagonal_product = 1
for i in range(4):
    diagonal_product *= matrix1[i][i]

print(f"Добуток елементів діагоналі першої матриці: {diagonal_product}")
```

```python
#15.Створіть матрицю розміром 5x5, заповніть її випадковими дійсними числами в діапазоні від -0.5 до 0.5 та знайдіть вектор-стовпчик, що містить кількість нульових елементів в кожному рядку матриці.
import numpy as np

# Створення матриці 5x5 з випадковими числами в діапазоні від -0.5 до 0.5
matrix = np.random.uniform(-0.5, 0.5, (5, 5))

# Знаходження вектора-стовпчика з кількістю нульових елементів в кожному рядку матриці
zero_counts = np.count_nonzero(matrix == 0, axis=1)

print("Матриця:")
print(matrix)
print("Вектор з кількістю нульових елементів в кожному рядку:")
print(zero_counts)
```
